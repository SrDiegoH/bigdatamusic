```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import spacy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def clear_text(text):
    nlp = spacy.load('en')
    doc = nlp(text)
    oputput = " ".join([token.lemma_ for token in doc if not token.is_stop])
    return oputput 
```

```python
MODELED = 'F:\\Sistema\\Downloads\\bigdatamusic_datas\\modeled\\'
LETTERS = f'{MODELED}letters.csv'
```

```python
#clear_text('thumbnail a e do gone tomas or first')
clear_text('i want to break free with my car')
# python -m spacy download en
```

```python
all_data = pd.read_csv(LETTERS, sep=';')
all_data.head()
```

```python
all_data = all_data.loc[:,['text']]

all_data['text'] = all_data['text'].apply(clear_text)

all_data = all_data.sample()

all_data.head()
#all_data
```

```python
#text = str(all_data['text'].to_string)
text = " ".join(music for music in all_data['text'])
text
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
